```python
pos = 'positive.txt'
neg = 'negative.txt'

negative = []
what = 0
with open(neg, 'r', encoding="utf-8",errors='ignore') as f:
    for line in f:
        if what == 0:
            pr = line.strip()
            what = 1
        else:
            seq = line.strip()
            what = 0
            negative.append([pr, seq])
       
positive = []
what = 0
with open(pos, 'r', encoding="utf-8",errors='ignore') as f:
    for line in f:
        if what == 0:
            pr = line.strip()
            what = 1
        else:
            seq = line.strip()
            what = 0
            positive.append([pr, seq])
print(len(positive))
print(len(negative))
```

```python
freq = {}
uniq_pos = []
for keys in positive:
    if freq.get(keys[1], -1) == -1:
        uniq_pos.append(keys)
    freq[keys[1]] = freq.get(keys[1], 0) + 1
freq_pos = freq
uniq_pos = np.array(uniq_pos)
print(uniq_pos.shape)
```

```python
freq = {}
uniq_neg = []
for keys in negative:
    if freq.get(keys[1], -1) == -1:
        uniq_neg.append(keys)
    freq[keys[1]] = freq.get(keys[1], 0) + 1
uniq_neg = np.array(uniq_neg)
print(uniq_neg.shape)
```

```python
uniq = []
for keys in uniq_neg:
    if freq_pos.get(keys[1], -1) == -1:
        uniq.append(keys)
uniq = np.array(uniq)
uniq_neg = uniq
print(uniq_neg.shape)
```
